Consider the situation when we are looking for change with $k=\$12$.  The Greedy algorithm will return $G(k) = 6$ for $\$12 = \$7 + \$1 + \$1 + \$1 + \$1 + \$1$, but an optimized algorithm $N(k)=3$ with the solution $\$12 = \$4 + \$4 + \$4$. Q.E.D

##### B. Give a recursive algorithm implementing the greedy algorithm of STEP A.

```python
def greedy(coins, selected, remain):
    if remain <= 0:
        return 
    
    for i in coins[::-1]:
        if i <= remain:
            selected.append(i)
            greedy(coins, selected, remain-i)
            return
    return
    
S=[]
C = [1, 4, 7, 13, 28, 52, 91, 365]
Goal = 25
greedy(C, S, Goal)
print(S)
```

##### F. Give a recursive top-down memoized dynamic programming algorithm implementing the recurrence of STEP B. Save whatever information you need to make the algorithm of STEP J possible.

```python
def wrapTopDown(money, goal):
    solutions = [0,1] + [0]*(goal)
    answer = recTopDown(solutions, money, goal)
    
    for index, i in enumerate(solutions):
        print(index, i)
    
    return answer

def recTopDown(solutions, money, remaining):
    if remaining <= 0:
        return solutions[0]
    
    m = []
    for i in [remaining - k for k in money if k <=remaining]:
        if solutions[i] == 0:
            m.append(recTopDown(solutions, money, i))            
    solutions[remaining] = min(m)+1
    return

goal = 10
money = [1, 4, 7, 13, 28, 52, 91, 365]

answer = wrapTopDown(money, goal)
print(money)
print(goal)
print(answer)
```
